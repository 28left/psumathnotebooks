# Try It Yourself

```python
def generator():
    
    f = function('quadratic')
    g = function('linear')
    
    answer_1 = sp.simplify(f(g(x)))
    answer_2 = sp.simplify(g(f(x)))
    
    statement = '''
        Let 
        $$f(x) = {}, \\qquad g(x) = {}$$ 
        '''.format(sp.latex(f(x)), sp.latex(g(x)))
    
    statement += r'''
        Determine the rule for the composite functions
        $$f \circ g \quad \text{and} \quad g \circ f$$
        '''    
    
    answer = r"""
        $$(f\circ g)(x) = {} \qquad (g \circ f)(x) = {}$$
        """.format(sp.latex(answer_1), sp.latex(answer_2))
    
    return statement, answer

    
P = BaseProblem("My Problem", 
                """
                <h2>Exercise 2:</h2>
                <hr>
                <p style="font-size:18px;">Determine the rule for the composite functions
                $$f \circ g \quad \\text{ and } \quad g \circ f$$</p>  
                """, 
                "expression", 
                "<b>Answer: </b>$$2^{n+1}-1$$")

R = GeneratorProblem("My Problem", 
                     "expression", 
                     generator)
```
