```python
%%makeproblem -wa 3450986_0
Question
<eqn>
A = randnum(-3,-1,1); 
AAA = -A*A*A;
B = randnum(1,3,1); BBB = B*B*B;
C = randnum(2,9,1); CB = C*B; CA = C*A;
D = randnum(6,15,3); DB = int(D*BBB/3); DA = int(D*AAA/3);
E = randnum(3,8,1); EB = E*B; EA = -E*A;
CONST = DB - EB + DA - EA;
err = DB + EB + DA + EA;
DBEB = DB - EB;
DAEA = DA - EA;
EQNS = [C, D, E, int(D/3)];
''</eqn>

<watex>
Evaluate \[\int_{$A}^{$B}{ e^{${C}x} + ${D}x^2 - $E dx}\]. &emsp;&emsp;&emsp;&emsp;&emsp;&emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp;  
</watex>

<_>

Answer
<watex>\[\dfrac{e^{$CB} - e^{$CA}}{$C} + $CONST\]</watex>
<watex>\[\dfrac{e^{$CB} - e^{$CA}}{$C} + $err\]</watex>
<watex>\[e^{$CB} - e^{$CA} + $CONST\]</watex>
<watex>\[e^{$CB} - e^{$CA} + $err\]</watex>
<watex>\[$C(e^{$CB} - e^{$CA}) + $err\]</watex>


Solution
<watex>
<center>\[
\begin{array}{rcll}
\int_{$A}^{$B}{ e^{$EQNS[0]x} + $EQNS[1]x^2 - $E dx} 
& = & \left(\dfrac{1}{$C}e^{$EQNS[0]x} + $D*\dfrac{1}{3}x^3 - $EQNS[2]x \right)\biggl|_{$A}^{$B} \\\
& = & \left(\dfrac{1}{$C}e^{$EQNS[0]x} + $EQNS[3]x^3 - $EQNS[2]x \right)\biggl|_{$A}^{$B} \\\
& = & \left(\dfrac{1}{$C}e^{$C*$B} + $EQNS[3]*$B^3 - $E*$B \right) - \left(\dfrac{1}{$C}e^{$C($A)} + $EQNS[3]($A)^3 - $E($A) \right) \\\
& = & \left(\dfrac{1}{$C}e^{$CB} + $EQNS[3]*$BBB - $EB \right) - \left(\dfrac{1}{$C}e^{$CA} - $EQNS[3]*$AAA + $EA \right) \\\
& = & \left(\dfrac{1}{$C}e^{$CB} + $DB - $EB \right) - \left(\dfrac{1}{$C}e^{$CA} - $DA + $EA \right) \\\
& = & \dfrac{1}{$C}e^{$CB} + $DBEB  - \dfrac{1}{$C}e^{$CA} + $DAEA  \\\
& = & \dfrac{e^{$CB} - e^{$CA}}{$C} + $CONST 
\end{array}
\]</center>
</watex>

Info
'tags': ["definite integral", "exponential function", "sum rule"],
'title': "Evaluate definite integral",
'id': "3450986_0"
```

```python
%%makeproblem -wa 3450986_1
Question
<eqn>
L = pickone(4,9,16); LL = int(sqrt(L)); LLL = LL**3; LLLLL = LL**5;
A = randnum(3,6,3); AA = int(2*A/3);
B = randnum(2,9,1,int(A/3)); BB = 2*B;
AALLL = AA*LLL; BBLL = BB*LL;
EQNS = [A, B, AA, BB];
ANSWER = AA*LLL - BB*LL - AA + BB;
WRONG1 = AA*LLL - BB*LL;
WRONG2 = AA*LLL - BB*LL - AA - BB;
WRONG3 = AA*LLL - BB*LL + AA - BB;
WRONG4 = AA*LLL - BB*LL + AA + BB;
''</eqn>

<watex>
Evaluate \[\int_{1}^{$L}{ \dfrac{${A}t - $B}{\sqrt{t}} dt}\]. &emsp;&emsp;&emsp;&emsp;&emsp;&emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp;  
</watex>

<_>

Answer
<watex><EQN $ANSWER></watex>
<watex><EQN $WRONG1></watex>
<watex><EQN $WRONG2></watex>
<watex><EQN $WRONG3></watex>
<watex><EQN $WRONG4></watex>

Solution
<watex>
<center>\[
\begin{array}{rcll}
\int_{1}^{$L}{ \dfrac{$EQNS[0]t - $B}{\sqrt{t}} dt} 
& = & \int_{1}^{$L}{\dfrac{$EQNS[0]t}{t^{1/2}} - \dfrac{$B}{t^{1/2}} dt} \\\ 
& = & \int_{1}^{$L}{$EQNS[0]t^{1/2} - $EQNS[1]t^{-1/2} dt} \\\ 
& = & \left($A*\dfrac{2}{3}t^{3/2} - $B*2t^{1/2}\right)\biggl|_{1}^{$L} \\\ 
& = & \left($EQNS[2]t^{3/2} - $EQNS[3]t^{1/2}\right)\biggl|_{1}^{$L} \\\ 
& = & \left($AA*$L^{3/2} - $BB*$L^{1/2}\right) - \left($AA*1^{3/2} - $BB*1^{1/2}\right) \\\ 
& = & $AA($L^{1/2})^3 - $BB*$L^{1/2} - $AA + $BB \\\ 
& = & $AA*$LL^3 - $BB*$LL - $AA + $BB \\\ 
& = & $AA*$LLL - $BB*$LL - $AA + $BB \\\  
& = & $AALLL - $BBLL - $AA + $BB \\\  
& = & $ANSWER 
\end{array}
\]</center>
</watex>

Info
'tags': ["definite integral", "root"],
'title': "Evaluate definite integral",
'id': "3450986_1"
```

### Problem 2

```python
%%makeproblem -wa 3444869_0
Question
<eqn>
C = randnum(2,6,1);
A = C * randnum(2,6,1); AC = int(A / C);
B = A + C * randnum(2,6,1); BC = int(B / C);
BCAC = BC - AC;
''</eqn>

<watex>
Find the area of the region under the graph of \[f(x) = e^{-x/$C}\] from \[x = $A\] to \[x = $B\]. 
</watex>

<_>

Answer
<watex>\[${C}e^{-$BC}(e^{$BCAC} - 1)\]</watex>
<watex>\[e^{-$BC}(e^{$BCAC} - 1)\]</watex>
<watex>\[${C}e^{$A}(e^{<eqn $B-$A>} - 1)\]</watex>
<watex>\[e^{$A}(e^{<eqn $B-$A>} - 1)\]</watex>
<watex>\[\dfrac{1}{$C}e^{$A}(e^{<eqn $B-$A>} - 1)\]</watex>

Solution
<watex>
First note that the area is given by the following definite integral
<br><br>
<center>\[\int_{$A}^{$B}{e^{-x/$C} dx}\]</center>
<br>
In order to integrate \[e^{-x/$C}\], use the following substitution:
<br><br>
<center>\[
\begin{array}{rcll}
u & = & -x/$C \\\ 
du & = &  -\dfrac{1}{$C} dx \qquad (-${C} du = dx) \\ 
\end{array}
\]</center>
<br><br>
<center>\[
\begin{array}{rcll}
\int_{x=$A}^{x=$B}{e^{-x/$C} dx} 
& = & \int_{u=-$A/$C}^{u=-$B/$C} e^u * (-$C) du \\\ 
& = & - $C\int_{u=-$AC}^{u=-$BC}{e^u du}  \\\
& = & - ${C}e^u\bigl|_{u=-$AC}^{u=-$BC} \\\ 
& = & - ${C}e^{-$BC} + ${C}e^{-$AC} \\\ 
& = & ${C}e^{-$BC}(e^{$BCAC} - 1)
\end{array}
\]</center>
</watex>

Info
'tags': ["integration", "area under graph"],
'title': "Find the area un the graph",
'id': "3444869_0"
```

### Problem 3

```python
%%makeproblem -wa 3440589_0
Question
<eqn>
X = 100*randnum(2,6,1);
if X/100 == 2:
    B = pickone(9,16,25,36,49,64,81,100); B = B - 4;
elif X/100 == 3: 
    B = pickone(16,25,36,49,64,81,100); B = B - 9;
elif X/100 == 4:
    B = pickone(25,36,49,64,81,100); B = B - 16;
elif X/100 == 5:
    B = pickone(36,49,64,81,100); B = B - 25;
else:
    B = pickone(49,64,81,100); B = B - 36;    
SQRT = sqrt( X*X/10000 + B);
A = int(100*SQRT*randnum(2,4,1));
P = int(10*randnum(4,9,1) + A/SQRT);
err = pick(4,-3,-2,-1,1,2,3,4);
''</eqn>

<watex>
Suppose the rate of change of the unit price \[p\] of video boards is
<br><br>
<center>\[p'(x) = -\dfrac{${A}x}{(x^2 + $B)^{3/2}}\]</center>
<br>
where \[x\] is the number of hundreds of video boards that the supplier will make available to the market daily and \[p\] is in dollars. Find the supply equation \[p(x)\] for the video boards if the quantity the supplier is willing to make available is $X video boards when the unit price is $P dollars. 
</watex>

<_>

Answer
<watex>\[\dfrac{$A}{\sqrt{x^2 + $B}} + <eqn int($P - $A/$SQRT)>\]</watex>
<watex>\[\dfrac{$A}{\sqrt{x^2 + $B}} + <eqn int($P - $A/$SQRT) + 10*$err[0]>\]</watex>
<watex>\[\dfrac{${A}x}{\sqrt{x^2 + $B}} + <eqn int($P - $A/$SQRT) + 10*$err[1]>\]</watex>
<watex>\[\dfrac{<eqn $A/2>x}{\sqrt{x^2 + $B}} + <eqn int($P - $A/$SQRT) + 10*$err[2]>\]</watex>
<watex>\[\dfrac{<eqn $A/2>}{\sqrt{x^2 + $B}} + <eqn int($P - $A/$SQRT) + 10*$err[3]>\]</watex>

Solution
<watex>
First find the general solution to the differential equation \[p'(x) =  -\dfrac{${A}x}{($B + x^2)^{3/2}}\].
<br><br>                                                               
<center>\[
\begin{array}{rcll}
p(x) & = & \int p'(x) dx \\\ 
& = & \int{-\dfrac{${A}x}{(x^2 + $B)^{3/2}} dx}&\\\ 
& = & -$A\int{x(x^2 + $B)^{-3/2} dx}&\\\ 
& = & -$A\int{(u)^{-3/2} \dfrac{1}{2} du} & \text{ using the substitution } u = x^2 + $B \quad du = 2x dx\\\ 
& = & -\dfrac{$A}{2}\int{(u)^{-3/2} du}\\\ 
& = & -\dfrac{$A}{2}(-2)u^{-1/2} + C\\\ 
& = & \dfrac{$A}{\sqrt{u}} + C\\\ 
& = & \dfrac{$A}{\sqrt{x^2 + $B}} + C
\end{array}
\]</center>
<br>
Therefore, \[p(x) = \dfrac{$A}{\sqrt{x^2 + $B}} + C\].  Now use the initial condition \[p(<eqn $X/100>) = $P\] to solve for \[C\].
<br><br>
<center>\[
\begin{array}{rcll}
$P & = & p(<eqn $X/100>) \\\
& = & \dfrac{$A}{\sqrt{(<eqn $X/100>)^2 + $B}} + C &\\\ 
& = & \dfrac{$A}{\sqrt{<eqn $X*$X/10000> + $B}} + C &\\\ 
& = & \dfrac{$A}{\sqrt{<eqn $X*$X/10000 + $B>}} + C &\\\ 
& = & \dfrac{$A}{<eqn int(sqrt($X*$X/10000 + $B))>} + C &\\\ 
& = & <eqn int($A/$SQRT)> + C &\\\ 
\end{array}
\]</center>
<br>
Therefore, \[C = $P - <eqn int($A/sqrt($X*$X/10000) + $B)> = <eqn int($P - $A/sqrt($X*$X/10000 + $B))>\] and 
<br><br>
<center>\[p(x) = \dfrac{$A}{\sqrt{x^2 + $B}} + <eqn int($P - $A/sqrt($X*$X/10000 + $B))>\]</center>
</watex>

Info
'tags': ["integration", "applications of integration", "differential equation", "supply equation"],
'title': "Supply Equation",
'id': "3440589_0"
```

```python
%%makeproblem -wa 4858132_0
Question
<eqn>
I = randnum(1,30,1);
[c,d] = pick(2,2,3,4,5,6,7);
A = randnum(2,5,1)*c;
B = A + randnum(2,5,1)*c;
a = int(A / c);
b = int(B / c);
EQNS = [c,d,dfraction(I,c),d*(b-a)];
ANSWER = fraction(I + d*(B - A),c);
DANSWER = dfraction(ANSWER);
ERR = pick(4,fraction(1,5),fraction(1,3),fraction(1,2),fraction(2,3),fraction(3,2),2,fraction(5,2));
WRONG1 = ERR[0]*ANSWER;
WRONG2 = ERR[1]*ANSWER;
WRONG3 = ERR[2]*ANSWER;
WRONG4 = ERR[3]*ANSWER;
''</eqn>

<watex>
If \[ \int_{$A}^{$B}{ f(x) dx}  = $I\], then compute the value of the following definite integral.
<br><br>
<center>\[ \int_{$a}^{$b}{f(${c}x) + $d \; dx}\]</center>
<br><br>
</watex>

<_>

Answer
<watex>\[<EQN dfraction($ANSWER)>\]</watex>
<watex>\[<EQN dfraction($WRONG1)>\]</watex>
<watex>\[<EQN dfraction($WRONG2)>\]</watex>
<watex>\[<EQN dfraction($WRONG3)>\]</watex>
<watex>\[<EQN dfraction($WRONG4)>\]</watex>

Solution
<watex>
Applying the sum rule for definite integrals, we have 
<br><br>
<center>\[
\int_{$a}^{$b}{f($EQNS[0]x) + $d dx} = \int_{$a}^{$b}{f($EQNS[0]x) dx} + \int_{$a}^{$b}{ $d dx}
\]</center>
<br>
To evaluate \[\int_{$a}^{$b}{f($EQNS[0]x) dx}\], use the following substitution:
<br><br>
<center>\[
\begin{array}{rcll}
u & = & $EQNS[0]x   & \\\
du & = & $EQNS[0] dx \qquad & \biggl(\dfrac{1}{$EQNS[0]}du = dx\biggr)
\end{array}
\]</center>
Therefore,
<br><br>
<center>\[
\begin{array}{rcll}
\int_{x=$a}^{x=$b}{f($EQNS[0]x) dx}
& = & \int_{u=$c\cdot$a}^{u=$c\cdot$b}{f(u) \dfrac{1}{$EQNS[0]}du} &\\\
& = &  \dfrac{1}{$EQNS[0]} \int_{u=$A}^{u=$B}{f(u) du} & \\\ 
& = &  $EQNS[2] & \text{ since } \int_{$A}^{$B}{ f(x) dx} = $I
\end{array}
\]</center>
<br>
And finally,
<br><br>
<center>\[
\begin{array}{rcl}
\int_{$a}^{$b}{f($EQNS[0]x) + $d dx} 
& = & \int_{$a}^{$b}{f($EQNS[0]x) dx} +  \int_{$a}^{$b}{ $d dx} \\\
& = & $EQNS[2] + $EQNS[1]x\biggl|_{$a}^{$b}\\\
& = & $EQNS[2] + $EQNS[1]($b - $a)\\\
& = & $EQNS[2] + $EQNS[3]\\\
& = & $DANSWER
\end{array}
\]</center>
</watex>

Info
'tags': ["integration", "substitution"],
'title': "Compute the integral",
'id': "4858132_0"
```
